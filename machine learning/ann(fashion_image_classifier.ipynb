```python
import tensorflow as tf
from tensorflow import keras
import numpy as np
import matplotlib.pyplot as plt
```

```python
fashion = keras.datasets.fashion_mnist
```

```python
(trainImages, trainLabels), (testImages, testLabels) = fashion.load_data()
```

```python
i = int(input("enter an index number"))
imag = trainImages[i]
plt.imshow(imag)
```

creating an neural network

```python
model = keras.Sequential([ keras.layers.Flatten(input_shape=(28,28)),
                          keras.layers.Dense(128, activation=tf.nn.relu),
                          keras.layers.Dense(10, activation=tf.nn.softmax)])
```

model compiler

```python
model.compile(optimizer = 'adam',
           loss = 'sparse_categorical_crossentropy',
           metrics=['accuracy'])
```

```python
model.fit(trainImages, trainLabels, epochs=5, batch_size=32)
```

```python
model.evaluate(testImages, testLabels)
```

```python
predictions = model.predict(testImages[0:5])
predictions
```
