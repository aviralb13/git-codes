```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('https://raw.githubusercontent.com/aviralb13/git-codes/main/datas/brainstroke.csv')
data.head()
```

```python
one_hot = pd.get_dummies(data['gender'])
data = data.drop('gender',axis = 1)
data = data.join(one_hot)

one_hot = pd.get_dummies(data['work_type'])
data = data.drop('work_type',axis = 1)
data = data.join(one_hot)

data.head()
```

```python
from sklearn import preprocessing

label = preprocessing.LabelEncoder()
data['ever_married'] = label.fit_transform(data['ever_married'])
data.head()
```

```python
import category_encoders

Residence_Mapping = [{'col':'Residence_type','mapping':{'Urban':0,'Rural':1}}]
smoke_Mapping = [{'col':'smoking_status','mapping':{'formerly smoked':1,'never smoked':2, 'smokes':0, 'Unknown':-1}}]

Residence_encoder = category_encoders.OrdinalEncoder(mapping = Residence_Mapping, return_df = True) 
smoke_encoder = category_encoders.OrdinalEncoder(mapping = smoke_Mapping, return_df = True) 

data['Residence_type'] = Residence_encoder.fit_transform(data['Residence_type'])
data['smoking_status'] = smoke_encoder.fit_transform(data['smoking_status'])
```

```python
data.head()
```

```python
features = ['age', 'hypertension', 'heart_disease', 'ever_married','Residence_type', 'avg_glucose_level', 'bmi', 'smoking_status','Female', 'Male', 'Govt_job', 'Private', 'Self-employed','children']
x = data[features]
y = data['stroke']
```

```python
from sklearn.model_selection import train_test_split

train_x,test_x,train_y,test_y = train_test_split(x,y)
```

```python
from lazypredict.Supervised import LazyClassifier

model = LazyClassifier()
models, predictions = model.fit(train_x,test_x,train_y,test_y)
models.head(60)
```

```python
from sklearn.ensemble import RandomForestClassifier
regressor_model = RandomForestClassifier()
regressor_model.fit(train_x,train_y) 
```

```python
predictions = regressor_model.predict(test_x)
```

```python
from sklearn.metrics import accuracy_score

accuracy_score(test_y, predictions)
```
