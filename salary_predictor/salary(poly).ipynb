```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv(r'/Users/apple/Documents/programming/datas/Position_Salaries.csv')
data.head()
```

```python
x = data.iloc[:, 1:-1].values
y = data.iloc[:, -1].values
```

```python
from sklearn.linear_model import LinearRegression
linear_model = LinearRegression()
linear_model.fit(x, y)
```

```python
from sklearn.preprocessing import PolynomialFeatures
polynomial_model = PolynomialFeatures(degree=4)
x_poly = polynomial_model.fit_transform(x)
linear_model2 = LinearRegression()
linear_model2.fit(x_poly,y)
```

```python
plt.scatter(x, y)
plt.plot(x, linear_model.predict(x))
```

```python
plt.scatter(x, y)
plt.plot(x, linear_model2.predict(x_poly))
```

```python
linear_model.predict([[8.5]])
```

```python
linear_model2.predict(polynomial_model.fit_transform([[8.5]]))
```
