```python
import category_encoders

webdev_Mapping = [{'col':'Did you do webdev during college time ?','mapping':{'no':0,'yes':1}}]
dataanalysis_Mapping = [{'col':'Are you good at Data analysis ?','mapping':{'no':0,'yes':1}}]
read_Mapping = [{'col':'reading and writing skills','mapping':{'poor':1,'medium':2,'excellent':3}}]
tech_Mapping = [{'col':'Are you a tech person ?','mapping':{'no':0,'yes':1}}]
soc_Mapping = [{'col':'Were you in a non tech society ?','mapping':{'no':0,'yes':1}}]
coding_Mapping = [{'col':'Are you good at coding ?','mapping':{'no':0,'yes':1}}]
app_Mapping = [{'col':'Have you developed mobile apps ?','mapping':{'no':0,'yes':1}}]
comm_Mapping = [{'col':'Are you good at communication ?','mapping':{'no':0,'yes':1}}]
security_Mapping = [{'col':'Do you have specialization in security','mapping':{'no':0,'yes':1}}]
data_Mapping = [{'col':'Have you ever handled large databases ?','mapping':{'no':0,'yes':1}}]
stats_Mapping = [{'col':'Do you have knowlege of statistics and data science?','mapping':{'no':0,'yes':1}}]
eng_Mapping = [{'col':'Are you proficient in English ?','mapping':{'no':0,'yes':1}}]
event_Mapping = [{'col':'Have you ever managed some event?','mapping':{'no':0,'yes':1}}]
blog_Mapping = [{'col':'Do you write technical blogs ?','mapping':{'no':0,'yes':1}}]
market_Mapping = [{'col':'Are you into marketing ?','mapping':{'no':0,'yes':1}}]
ml_Mapping = [{'col':'Are you a ML expert ?','mapping':{'no':0,'yes':1}}]
connections_Mapping = [{'col':'Do you have a lot of connections ?','mapping':{'no':0,'yes':1}}]
liveproject_Mapping = [{'col':'Have you ever built live project ?','mapping':{'no':0,'yes':1}}]

webdev_encoder = category_encoders.OrdinalEncoder(mapping = webdev_Mapping, return_df = True)  
dataanalysis_encoder = category_encoders.OrdinalEncoder(mapping = dataanalysis_Mapping, return_df = True)  
read_encoder = category_encoders.OrdinalEncoder(mapping = read_Mapping, return_df = True)  
tech_encoder = category_encoders.OrdinalEncoder(mapping = tech_Mapping, return_df = True)  
soc_encoder = category_encoders.OrdinalEncoder(mapping = soc_Mapping, return_df = True)  
coding_encoder = category_encoders.OrdinalEncoder(mapping = coding_Mapping, return_df = True)  
app_encoder = category_encoders.OrdinalEncoder(mapping = app_Mapping, return_df = True)  
comm_encoder = category_encoders.OrdinalEncoder(mapping = comm_Mapping, return_df = True)  
security_encoder = category_encoders.OrdinalEncoder(mapping = security_Mapping, return_df = True)  
data_encoder = category_encoders.OrdinalEncoder(mapping = data_Mapping, return_df = True)  
stats_encoder = category_encoders.OrdinalEncoder(mapping = stats_Mapping, return_df = True)  
eng_encoder = category_encoders.OrdinalEncoder(mapping = eng_Mapping, return_df = True)  
event_encoder = category_encoders.OrdinalEncoder(mapping = event_Mapping, return_df = True)  
blog_encoder = category_encoders.OrdinalEncoder(mapping = blog_Mapping, return_df = True)  
market_encoder = category_encoders.OrdinalEncoder(mapping = market_Mapping, return_df = True)  
ml_encoder = category_encoders.OrdinalEncoder(mapping = ml_Mapping, return_df = True)  
connections_encoder = category_encoders.OrdinalEncoder(mapping = connections_Mapping, return_df = True)  
liveproject_encoder = category_encoders.OrdinalEncoder(mapping = liveproject_Mapping, return_df = True)  

data['Did you do webdev during college time ?'] = webdev_encoder.fit_transform(data['Did you do webdev during college time ?'])
data['Are you good at Data analysis ?'] = dataanalysis_encoder.fit_transform(data['Are you good at Data analysis ?'])
data['reading and writing skills'] = read_encoder.fit_transform(data['reading and writing skills'])
data['Are you a tech person ?'] = tech_encoder.fit_transform(data['Are you a tech person ?'])
data['Were you in a non tech society ?'] = soc_encoder.fit_transform(data['Were you in a non tech society ?'])
data['Are you good at coding ?'] = coding_encoder.fit_transform(data['Are you good at coding ?'])
data['Have you developed mobile apps ?'] = app_encoder.fit_transform(data['Have you developed mobile apps ?'])
data['Are you good at communication ?'] = comm_encoder.fit_transform(data['Are you good at communication ?'])
data['Do you have specialization in security'] = security_encoder.fit_transform(data['Do you have specialization in security'])
data['Have you ever handled large databases ?'] = data_encoder.fit_transform(data['Have you ever handled large databases ?'])
data['Do you have knowlege of statistics and data science?'] = stats_encoder.fit_transform(data['Do you have knowlege of statistics and data science?'])
data['Are you proficient in English ?'] = eng_encoder.fit_transform(data['Are you proficient in English ?'])
data['Have you ever managed some event?'] = event_encoder.fit_transform(data['Have you ever managed some event?'])
data['Do you write technical blogs ?'] = blog_encoder.fit_transform(data['Do you write technical blogs ?'])
data['Are you into marketing ?'] = market_encoder.fit_transform(data['Are you into marketing ?'])
data['Are you a ML expert ?'] = ml_encoder.fit_transform(data['Are you a ML expert ?'])
data['Do you have a lot of connections ?'] = connections_encoder.fit_transform(data['Do you have a lot of connections ?'])
data['Have you ever built live project ?'] = liveproject_encoder.fit_transform(data['Have you ever built live project ?'])
```
