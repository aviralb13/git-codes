```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = pd.read_csv(r"/Users/apple/Documents/programming/datas/shelflife2.csv")
data.head()
```

```python
z = data.iloc[: , -2].values
a = data.iloc[: , 0:-2].values 
b = data.iloc[: , -1].values 
c = data.iloc[:,0].values
d = data.iloc[:,1].values

temperature = data.iloc[: , 0].values
moisture = data.iloc[: , 1].values
days = data.iloc[:, 2].values
```
